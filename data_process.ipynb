```python
df["vanilla"] = ''
df["CoT"] = ''
df["self_refinement"] = ''
df["debate"] = ''
df["MADR"] = ''
```

```python
df = vanilla(df)
```

```python
df = CoT(df)
```

```python
df = self_refinement(df)
```

```python
df = debate(df)
```

```python
df = madr(df)
```
